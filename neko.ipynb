```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cats_dataset.csv')
data
```

```python
data.columns
```

```python
data['Age (Years)'].plot.hist(bins=5, alpha=0.7, color='#5226CB')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Distribusi umur paling banyak terdapat di umur sekitar 17.5 tahun, sedangkan distribusi umur paling sedikit berada di sekitar 10 tahun.

```python
breed_counts = data['Breed'].value_counts()

breed_counts.plot(kind='bar', figsize=(10, 6), color='magenta', edgecolor='white')

plt.title('Frequency of Each Cat Breed')
plt.xlabel('Breed')
plt.ylabel('Count')

plt.xticks(rotation=50, ha='right')

plt.tight_layout()
plt.show()
```

Jumlah kucing paling banyak adalah Ragdoll, sedangkan jumlah kucing yang paling sedikit adalah Norwegian Forest.

```python
breed_counts = data['Weight (kg)'].value_counts()

breed_counts.plot(kind='bar', figsize=(10,6), color='blue', edgecolor='white')

plt.title('Frequency of Weight (kg) Cat Breed')
plt.xlabel('Weight (kg)')
plt.ylabel('Count')

plt.xticks(rotation=50, ha='right')

plt.tight_layout()
plt.show()
```

Berat kucing mayoritas sebesar 5 kg, sedangkan berat kucing minoritas adalah 2 kg.

```python
data['Weight (kg)'].plot.hist(bins=4, alpha=0.98, color='#FFCDB8')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

Dari tabel histogram, nampak berat kucing terbanyak berada di 5 kg. Sedangkan berat kucing tersedikit berada di 2 kg.

```python
breed_counts = data['Color'].value_counts()

breed_counts.plot(kind='bar', figsize=(9, 4), color='violet')

plt.title('Frequency of Color Each Cat Breed')
plt.xlabel('Color')
plt.ylabel('Count')

plt.xticks(rotation=50, ha='right')

plt.tight_layout()
plt.show()
```

Tricolor adalah warna terbanyak yang ada di bulu kucing pada data, sedangkan white adalah warna tersedikit yang ada pada kucing pada data.

```python
breed_counts = data['Gender'].value_counts()

breed_counts.plot(kind='bar', figsize=(10, 6), color='yellow')

plt.title('Gender of Cat Breed')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.xticks(rotation=50, ha='right')

plt.tight_layout()
plt.show()
```

Gender terbanyak pada data kucing adalah male, sedangkan female berbeda tipis dengan male.
